```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import tensorflow.keras.layers as layers
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
Data = np.loadtxt("nd5table.txt")

X = np.transpose(Data[:,:25].reshape((-1,5,5,1)),(0,2,1,3))
Y = Data[:,-2]
plt.hist(Y)
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33)
X[np.random.randint(0,X.shape[0],1),:,:,0], X_train.shape
```

```python
tf.keras.backend.clear_session()

dim = 5; dr = 0.2
model = keras.Sequential([
            layers.Conv2D(64, (3,3), activation='relu', padding='same', input_shape=(dim,dim,1)),
            layers.Conv2D(32, (3,3), activation='relu', padding='same'),
            layers.MaxPool2D((2,2)),
            layers.Dropout(dr),
            layers.Conv2D(64, (2,2), activation='relu'),
            layers.Flatten(),
            layers.BatchNormalization(),
            layers.Dense(1024, activation='relu'),
            layers.Dropout(dr),
            layers.Dense(128, activation='relu'),
            layers.Dense(1)
        ])

model.compile(optimizer=tf.keras.optimizers.Adam(lr=1e-3),
              loss='mse')
# mse; binary_crossentropy; kullback_leibler_divergence
model.summary()
```

```python
model.fit(X_train,-np.log((1+1e-3)/y_train-1), validation_split=0.2,
          batch_size=2048, epochs=128)
```

```python
pred = model.predict(X_test)
pred = pred.reshape((-1,))
pred = 1 / (1 + np.exp(-pred))

print(np.mean((y_test<0.4) == (pred<0.4)))

f = plt.figure(figsize=(12,3))
ax1 = f.add_subplot(131)
ax2 = f.add_subplot(132)
ax3 = f.add_subplot(133)

ax1.hist(pred)
ax1.title.set_text("ptrue_pred hist")

ax2.scatter((y_test+pred)/2, (pred-y_test), alpha = 0.1)
ax2.title.set_text("ptrue-pred ~ (ptrue+pred)/2")

ax3.hist(np.abs(y_test-pred))
ax3.title.set_text("hist of abs pred - ptrue")

plt.show()
```
